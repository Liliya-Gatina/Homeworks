```python
def Minmax(X, Y):
    if X > Y:
        X, Y = Y, X
    return X, Y

A = float(input('A='))
B = float(input('B='))
C = float(input('C='))
D = float(input('D='))

A, B = Minmax(A, B)
C, D = Minmax(C, D)
A, C = Minmax(A, C)
B, D = Minmax(B, D)

print(f'Min: {A}, Max: {D}')
```

```python
def SortDec3(A, B, C):
    if A < B:
        A, B = B, A
    if B < C:
        B, C = C, B
    if A < B:
        A, B = B, A
    return A, B, C

A1 = float(input('A1='))
B1 = float(input('B1='))
C1 = float(input('C1='))
A2 = float(input('A2='))
B2 = float(input('B2='))
C2 = float(input('C2='))

A1, B1, C1 = SortDec3(A1, B1, C1)
A2, B2, C2 = SortDec3(A2, B2, C2)

print(A1, B1, C1)
print(A2, B2, C2)
```

```python
def SumRange(A, B):
    if A > B:
        return 0
    return sum(range(A, B + 1))

A = int(input())
B = int(input())
C = int(input())

sum_AB = SumRange(A, B)
sum_BC = SumRange(B, C)

print(sum_AB)
print(sum_BC)
```

```python
def Quarter(x, y):
    if x > 0 and y > 0:
        return 1
    elif x < 0 and y > 0:
        return 2
    elif x < 0 and y < 0:
        return 3
    else:
        return 4

for _ in range(3):
    x = float(input("X="))
    y = float(input("Y="))
    quarter = Quarter(x, y)
    print(f"Точка ({x}, {y}) находится в {quarter}-й четверти.")
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
```

```python

def find_twin_primes(n):
    result = []
    for i in range(n, 2 * n + 1):
        if is_prime(i) and is_prime(i + 2):
            result.append((i, i + 2))
    return result
```

```python
n = 10
print(find_twin_primes(n))
```

```python
import math

def distance(p1, p2):
    return math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)
```

```python
def max_distance(points):
    max_dist = 0
    pair = None

    for i in range(len(points)):
        for j in range(i+1, len(points)):
            dist = distance(points[i], points[j])
            if dist > max_dist:
                max_dist = dist
                pair = (points[i], points[j])

    return pair, max_dist
```

```python
points = [(1, 2), (4, 6), (5, 7)]
result = max_distance(points)
print(f'Максимальное расстояние между точками {result[0]}, оно равно {result[1]}')
```

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)
```

```python
a = 48
b = 180
print(gcd(a, b))
```
