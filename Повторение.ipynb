**Группа While**

1. image.png

```python
N = int(input())
K = int(input())
count = 0
while N >= K:
    N -= K
    count += 1
print(count)
print(N)
```

3. image.png

```python
N = int(input())
count = 0
sum = 0
while N > 0:
    sum += N % 10
    count += 1
    N //= 10
print(count)
print(sum)
```

4. image.png

```python
N = int(input())
i = 2
prime = True
while i * i <= N and prime:
    if N % i == 0:
        prime = False
    i += 1
print(prime)
```

5. image.png

```python
A = int(input())
B = int(input())
C = int(input())
count = 0
rows = 0
while rows * C < A:
    cols = 0
    while cols * C < B:
        count += 1
        cols += 1
    rows += 1
print(count)
```
