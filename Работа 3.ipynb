```python
for A in [True, False]:
    for B in [True, False]:
        for C in [True, False]:
            print(f"A={A}, B={B}, C={C}")

            expression_1a = A or B
            expression_1b = B or A
            if expression_1a != expression_1b:
                print("Ошибка! Первая пара выражений не эквивалентна.")
            else:
                print("Первая пара выражений эквивалентна.")

            expression_2a = A and B
            expression_2b = B and A
            if expression_2a != expression_2b:
                print("Ошибка! Вторая пара выражений не эквивалентна.")
            else:
                print("Вторая пара выражений эквивалентна.")

            expression_3a = (A or B) or C
            expression_3b = A or C
            if expression_3a != expression_3b:
                print("Ошибка! Третья пара выражений не эквивалентна.")
            else:
                print("Третья пара выражений эквивалентна.")

            expression_4a = (A and B) and C
            expression_4b = A and (B and C)
            if expression_4a != expression_4b:
                print("Ошибка! Четвертая пара выражений не эквивалентна.")
            else:
                print("Четвертая пара выражений эквивалентна.")

            expression_5a = A and (A or B)
            expression_5b = A
            if expression_5a != expression_5b:
                print("Ошибка! Пятая пара выражений не эквивалентна.")
            else:
                print("Пятая пара выражений эквивалентна.")

            expression_6a = A or (A and B)
            expression_6b = A
            if expression_6a != expression_6b:
                print("Ошибка! Шестая пара выражений не эквивалентна.")
            else:
                print("Шестая пара выражений эквивалентна.")

            expression_7a = A and (B or C)
            expression_7b = (A and B) or (A and C)
            if expression_7a != expression_7b:
                print("Ошибка! Седьмая пара выражений не эквивалентна.")
            else:
                print("Седьмая пара выражений эквивалентна.")

            expression_8a = A or (B and C)
            expression_8b = (A or B) and (A or C)
            if expression_8a != expression_8b:
                print("Ошибка! Восьмая пара выражений не эквивалентна.")
            else:
                print("Восьмая пара выражений эквивалентна.")
```

```python
import math

epsilon = 1e-3
sum_series = 0.0
n = 1
a_n = math.factorial(n) / (n ** n)

while a_n >= epsilon:
    sum_series += a_n
    n += 1
    a_n = math.factorial(n) / (n ** n)

sum_series += a_n
print(sum_series)
```
