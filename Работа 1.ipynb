```python
import math

def rad_to_dms(radians):
    degrees = int(math.degrees(radians))
    minutes = int((math.degrees(radians) - degrees) * 60)
    seconds = (math.degrees(radians) - degrees - minutes / 60) * 3600
    return degrees, minutes, seconds

radians = float(input("Угол: "))
d, m, s = rad_to_dms(radians)
print(f"{d}° {m}' {s:.2f}''")
```
